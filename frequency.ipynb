```python
import string
from collections import Counter
import random

# Alfabeto español
alphabet = list(string.ascii_uppercase) + ['Ñ']

def caesar_encrypt(text, displacement):
    encrypted_text = ''
    for char in text.upper():
        if char in alphabet:
            new_char_index = (alphabet.index(char) + displacement) % len(alphabet)
            encrypted_text += alphabet[new_char_index]
        else:
            encrypted_text += char
    return encrypted_text

def caesar_decrypt(cyphered_text, displacement):
    decrypted_text = ''
    for char in cyphered_text.upper():
        if char in alphabet:
            original_char_index = (alphabet.index(char) - displacement) % len(alphabet)
            decrypted_text += alphabet[original_char_index]
        else:
            decrypted_text += char
    return decrypted_text


# Función para calcular la distribución de caracteres en el texto cifrado
def calculate_distribution(cyphered_text):
    # Contar la frecuencia de cada carácter en el texto cifrado
    char_count = Counter(cyphered_text.upper())
    total_chars = len(cyphered_text)
    distribution = {}
    
    # Calcular la probabilidad de cada carácter en el alfabeto español
    for char in alphabet:
        frequency = char_count[char] / total_chars if char_count[char] > 0 else 0
        distribution[char] = frequency
    
    return distribution

# Función para generar un texto cifrado aleatorio con distribución variada
def generate_random_cyphered_text():
    # Crear una lista con todas las letras del alfabeto español
    all_chars = list(alphabet)
    
    # Crear una lista para almacenar las letras del texto cifrado
    cyphered_chars = []
    
    # Generar un número aleatorio de caracteres para el texto cifrado
    num_chars = random.randint(50, 100)
    
    # Generar el texto cifrado aleatorio con una distribución variada
    for _ in range(num_chars):
        # Elegir aleatoriamente una letra del alfabeto español
        char = random.choice(all_chars)
        # Agregar la letra al texto cifrado
        cyphered_chars.append(char)
        # Eliminar la letra de la lista para evitar repeticiones
        all_chars.remove(char)
        # Si ya se han usado todas las letras, reiniciar la lista
        if len(all_chars) == 0:
            all_chars = list(alphabet)
    
    # Convertir la lista de caracteres en una cadena y devolverla
    return ''.join(cyphered_chars)
```

```python
# Generar un texto cifrado aleatorio que incluya todas las letras
cyphered_text_random = generate_random_cyphered_text()

# Imprimir el texto cifrado aleatorio
print("Texto cifrado aleatorio:")
print(cyphered_text_random)

# Calcular la distribución en el texto cifrado aleatorio
distribution_random = calculate_distribution(cyphered_text_random)

# Imprimir la distribución en el texto cifrado aleatorio
print("\nDistribución en el texto cifrado aleatorio:")
for char, prob in distribution_random.items():
    print(f"{char}: {prob:.2f}")
```

```python
import matplotlib.pyplot as plt
# Graficar la distribución de caracteres en el texto cifrado aleatorio
plt.bar(distribution_random.keys(), distribution_random.values())
plt.xlabel('Caracteres')
plt.ylabel('Probabilidad')
plt.title('Distribución de caracteres en el texto cifrado aleatorio')
plt.show()
```

```python
# Distribución teórica de frecuencias de las letras en español
theoretical_distribution = {
    'A': 0.1253, 'B': 0.0142, 'C': 0.0468, 'D': 0.0586, 'E': 0.1368,
    'F': 0.0069, 'G': 0.0101, 'H': 0.0070, 'I': 0.0625, 'J': 0.0044,
    'K': 0.0002, 'L': 0.0497, 'M': 0.0315, 'N': 0.0671, 'Ñ': 0.0031,
    'O': 0.0868, 'P': 0.0251, 'Q': 0.0088, 'R': 0.0684, 'S': 0.0798,
    'T': 0.0463, 'U': 0.0291, 'V': 0.0111, 'W': 0.0008, 'X': 0.0021,
    'Y': 0.0108, 'Z': 0.0040
}

# Función para comparar la distribución encontrada con la distribución teórica
def compare_distributions(found_distribution, theoretical_distribution):
    comparison_results = {}
    for char in alphabet:
        theoretical_prob = theoretical_distribution[char]
        found_prob = found_distribution[char] if char in found_distribution else 0
        comparison_results[char] = (theoretical_prob, found_prob)
    return comparison_results

# Ejemplo de uso:
# Calcular la comparación de distribuciones
comparison_results = compare_distributions(distribution_random, theoretical_distribution)

# Imprimir los resultados de la comparación
print("Comparación de distribuciones encontrada y teórica:")
for char, (theoretical_prob, found_prob) in comparison_results.items():
    print(f"{char}: Teórica={theoretical_prob:.4f}, Encontrada={found_prob:.4f}")
```

```python
import numpy as np

# Función para graficar la comparación de distribuciones
def plot_distribution_comparison(comparison_results):
    chars = comparison_results.keys()
    theoretical_probs = [theoretical_prob for theoretical_prob, _ in comparison_results.values()]
    found_probs = [found_prob for _, found_prob in comparison_results.values()]

    x = np.arange(len(chars))
    width = 0.35

    fig, ax = plt.subplots()
    rects1 = ax.bar(x - width/2, theoretical_probs, width, label='Teórica')
    rects2 = ax.bar(x + width/2, found_probs, width, label='Encontrada')

    ax.set_xlabel('Caracteres')
    ax.set_ylabel('Probabilidad')
    ax.set_title('Comparación de distribuciones teórica y encontrada')
    ax.set_xticks(x)
    ax.set_xticklabels(chars)
    ax.legend()

    fig.tight_layout()

    plt.show()

# Ejemplo de uso:
# Graficar la comparación de distribuciones
plot_distribution_comparison(comparison_results)
```
